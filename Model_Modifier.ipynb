```python
def filter_results(m, seasons, number):
    newd = {}
    for s in seasons:
        ndf = m.results[s]
        preds = ndf['PREDICTION']
        spreads = ndf['SPREAD']
        fl = []
        for i in range(len(preds)):
            if abs(preds[i] + spreads[i]) >= number:
                fl.append(True)
            else:
                fl.append(False)
        newd[s] = ndf[fl]
    return newd
        
    
```

```python
for poopoo in range(30):
    print(f"THRESHHOLD: {poopoo}")
    newd = filter_results(m, SEASONS, poopoo)
    for s in SEASONS:
        df699 = newd[s]
        acc = df699['HIT'].value_counts()[2] / (df699['HIT'].value_counts()[0] + df699['HIT'].value_counts()[2])
        print(f"accuracy for {s} is {acc})")
```
